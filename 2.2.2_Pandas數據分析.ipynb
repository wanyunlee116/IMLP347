## 4 增加一行

### 【技巧】

我們增加一行, 加入總級分。

```python
df['總級分'] = df.sum(axis=1) #🌟重複執行總級分會再次加總 因為axis又變成ANS的數值了
#data需重新跑一次
df.head()
```

### 【技巧】

有計算的當然也可以的。

```python
df['加權'] = df.國文 + df.數學*2 + df.英文
```

```python
df.head()
```

## 5 排序和 index 重設

### 【重點】排序的方法

我們依總級分來排序。

```python
df.sort_values(by='總級分', ascending=False).head(10)
```

### 【重點】排序的方法

加權分最高, 同分才看總級分

```python
df2 = df.sort_values(by=['加權', '總級分'], ascending=False)
df2.head(10)
```

### 【重點】重設 index

```python
#重設index值 方便之後畫圖
df2.index = range(1,101) #不含終止值 
df2.head()
```

## 6 篩出我們要的資料

基本上和 NumPy 的 array 篩法很像。

### 【重點】

找出數學滿級分同學。

```python
df2[df2.數學==15] #🌟'='是assign!!!
```

### 【重點】

找出數學和英文都滿級分的同學。要注意 `and` 要用 `&`, `or` 要用 `|`。每個條件一定要加弧號。

```python
df2[(df2.數學==15) & (df2.英文==15)] #or: df2[(df2.數學==15) | (df2.英文==15)]
```

## 9 手工打造一個 DataFrame*

有時我們用手工打造一個簡單的 DataFrame, 可以更理解整個結構。其實很容易, 一個 DataFrame 基本上就包含兩個主要部份:

* 資料本身: 通常一個二維陣列 (矩陣)
* 行、列的名稱

我們來個簡單的小例子。

```python
#記得 import numpy as np
#記得 import marplotlib.pyplot as plt
mydata = np.random.randn(4,3)
mydata
```

把行列的名字放進去, 就成一個 DataFrame。我們列的部份先讓 Python 自己產生。

```python
df2 = pd.DataFrame(mydata, columns=list("ABC")) #整個列表變成DataFrame了
```

```python
df2
```

#### 兩個表格上下貼起來

我們再來生一個 DataFrame, 再「貼」起來。

```python
df3 = pd.DataFrame(np.random.randn(3,3), columns=list("ABC")) #mydata直接改
```

```python
df3
```

```python
#Q：兩上表上下疊起來 
df4 = pd.concat([df2, df3], axis=0) #concatenate:連接 #設定根據哪個軸concat, 上下:axis=0
```

```python
df4
```

前面我們弄得亂七八糟的 index 重設一下。

```python
df4.index = range(len(df4)) #or range(7)
```

```python
df4
```

#### 横向的貼

```python
df5 = pd.concat([df2, df3], axis=1) #改axis=1
```

等等, 這大小好像不太對也可以嗎? 答案是可以的!

```python
df5
```
