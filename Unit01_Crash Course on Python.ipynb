#### 請用戶輸入 Email 地址，然後判斷它是自定義域名還是熱門域名。

> 題目：<br>
輸入(Input):shelly200318@hotmail.com.tw  <br>
輸出(Output):Your username is 'shelly200318' and your domain name is 'hotmail.com.tw' <br>


* 進階：把常用的信箱存成字典，加入判斷是否為
>輸入(Input):mary.jane@gmail.com  <br>
輸出(Output):這是註冊在 Google 之下的 Email 地址 <br>
輸入(Input):matt.pan@myfantasy.com  <br>
輸出(Output):這是在 myfantasy 之下自定義域



````python
# Get user email address
email = input('What is your email address?: ')#.strip()#移除字串前後有的空格e.g. ' ABC '
#bobo2c07@gmail.com

# Slice out the user name（@前面的字）
#所有字串都是有順序性so可以用[]來做特定位置擷取
user_name = email[0:email.index('@')] #0可省略 #@index的位置

# Slice the domain name（@後面的域名）
domain_name = email[email.index('@')+1:len(email)]#？？why m not -1 #到字串的最後一個位置，也可以不用放[OOO:]

# Format message
output = "Your username is '{}' and your domain name is '{}'".format(user_name, domain_name)#🌟
print(output)


#----------進階題----------#
domain = {'gmail.com':'Google','yahoo.com.tw':'Yahoo','ntu.edu.tw':'臺大','hotmail.com.tw':'Hotmail','hotmail.com':'Hotmail'}
if domain_name in domain:
    output = "這是註冊在 {} 之下的 Email 地址".format(domain[domain_name]) #查domain_name在domain字典的哪個🌟位置[]
else:
    output = "這是在 {} 之下自定義域".format(domain[domain_name])
print(output)
````

### 08 - 資料夾管理(shutil)
#### 說明 ####

```
在目前所在的目錄下建立一files資料夾

令使用者輸入一數字N，並在files資料夾中建立f1, f2… fN等N個資料夾後列出files的資料夾內容

將files資料夾裡的f1資料夾重新命名成folder1後再列出files的資料夾內容

移除files資料夾中的folder1後再列出files的資料夾內容

最後移除files資料夾
※須先退出files資料夾(os.chdir(../)) 
```


````python
# 以為ML時方便管理

import os
import shutil

if os.path.exists('files'): #確認目前路徑（同本支程式）有沒有files 並建立（不用像第7題還要輸路徑）
    shutil.rmtree('files') #shell utility:在Py中實現文件複製、移動、壓縮、解壓等高級功能(高階文件操作的函式庫) #Remove-Tree 目錄樹 #if目前路徑有files：移除
os.mkdir('files') #make directory 新增一files

n = int(input()) #請使用者輸入n, 輸多少就建立幾個資料夾
os.chdir('files') #check directory改到files路徑。同一層ㄉ完整寫法：'./files'
for i in range(1, n+1): #進files裡面一個一個生成 #(從1開始, 終止值不包含所以要+1)
    os.mkdir('f'+str(i)) #因為題目是f1, f2... #i是int 要轉成string才能做檔案命名
    print(sorted(os.listdir())) #對 list 中的元素進行排序 對元素排序之後，傳回一個排序好的新 list，而原本的 list 則不受影響; sort:會直接改變原本的 list 內容
a = input('Enter') #停頓 沒有使用到a變數so 意思是請使用者按Enter才會繼續執行再存成a #沒停頓的話 執行很快會看不到結果 因為後面會把files移除 題目倒數第二行）#需degug可這樣做

os.rename('f1', 'folder1')
print(sorted(os.listdir()))
a = input('Enter') 

os.rmdir('folder1')
print(sorted(os.listdir()))
a = input('Enter')  

os.chdir('../') #要移除folder要先回「上一層路徑」==> ../ ；同一層==> ./ 或省略寫
shutil.rmtree('files')
````

````python
#練習自己定義函示
#程式由上往下讀會知道有4個def自定義函示 但不會執行各項的內容 直到呼叫函示🌟

def convertInputString(): #def:自己定義的函示, return:回傳值
    #請使用者輸入一串字串，先轉換字串為統一小寫
    rawInput = input("\nPlease enter a word, phrase, or a sentence \nto check if it is a palindrome: ") 
    rawString = rawInput.lower() #因為是英文 要先把大小寫都統一 這裡是 變小寫lower（中文不影響）
    rawList = list(rawString)
    return rawList

def stripAnalphabetics(dirtyList): #dirtyList:現在還有很多不必要的符號
    analphabeticList = [" ", "-", ".", ",", ":", ";", "!", "?", "'", "\""] #生成一個list
    for character in analphabeticList: #for迴圈走訪每個標點符號
        if character in dirtyList:
            dirtyList.remove(character) #跟上面replace（用於string）意思一樣 🌟但list要用rm #rm完那些character（標點符號）後
            return stripAnalphabetics(dirtyList) 
    #去除相關標點符號
    return dirtyList #if已經都沒有標點符號了（對if不會是True不會進裡面） 現在的是乾淨的list
   
def runPalindromeCheck(straightList): #straightList是區域變數, 在main主程式的 引數 originalList傳回來會是 參數 straightList
    reversedList = straightList[::-1] #回文正反看都一樣 #::-1 means reverse反過來*以下示範*
    if reversedList == straightList: #if兩者相等 ==> 就會是回文 
        return "The text you have entered is a palindrome!" 
    else: 
        return "The text you have entered is not a palindrome." 

def main(): #main:主程式
    print("\nPalindrome checker") 
    originalList = convertInputString()
    originalList = stripAnalphabetics(originalList) #去除標點符號
    palindromeCheck = runPalindromeCheck(originalList)
    print(palindromeCheck)

main() #函示呼叫方式 呼叫後才會進到函示第二行開始往下執行🌟 if有設定參數（e.g.第二個def）這裡要給引數
````
